### Reading Pearson's Files

```python
import pandas as pd
import numpy as np


import plotly
import plotly.plotly as py
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from plotly.graph_objs import *

init_notebook_mode(connected=True)
```

This notebook reads and plots the output of the script `Sample_size_analysis.py`. Note that the number of PCs to retain (an argument of that script) is included in the name of the file, as is the choice by the user of Scaling matrices pior to PCA or to multiply its eigenvectors by their respective eigenvalues.

```python
Eigen= False
Scale= False
PC= 8

File= 'Sampling_test_pearsons_' + str(PC) + 'PC_Eigen='+ str(Eigen) +'_Scaled='+ str(Scale) +'.txt'

tests= []


Complete= open(File,'r')

d= 0
for line in Complete:
    line= line.split()
    if d== 0:
        Names= line
        d += 1
        continue
    
    line[:2]= [int(x) for x in line[:2]]
    line[2:]= [float(x) for x in line[2:]]
    tests.append(line)

Complete.close()

tests= np.array(tests)
```

```python
#############
############# PC-wise Pearson's

### Biased Pearsons
File= 'Sampling_BIAS_PC-wise_pears_' + str(PC) + 'PC_Eigen='+ str(Eigen) +'_Scaled='+ str(Scale) +'.txt'

Bias_p= []


Complete= open(File,'r')

d= 0
for line in Complete:
    line= line.split()
    if d== 0:
        Names= line
        d += 1
        continue
    
    line[:2]= [int(x) for x in line[:2]]
    line[2:]= [float(x) for x in line[2:]]
    Bias_p.append(line)

Complete.close()

Bias_p= np.array(Bias_p)

### MScorrected Pearsons
File= 'Sampling_MScorr_PC-wise_pears_' + str(PC) + 'PC_Eigen='+ str(Eigen) +'_Scaled='+ str(Scale) +'.txt'

MScorr= []


Complete= open(File,'r')

d= 0
for line in Complete:
    line= line.split()
    if d== 0:
        Names= line
        d += 1
        continue
    
    line[:2]= [int(x) for x in line[:2]]
    line[2:]= [float(x) for x in line[2:]]
    MScorr.append(line)

Complete.close()

MScorr= np.array(MScorr)
```

```python
N_pop= 4

labels1= {z:[x for x in range(Bias_p.shape[0]) if Bias_p[x,1] == z] for z in list(set(Bias_p[:,1]))}

fig_data= [go.Scatter(
        x = [Bias_p[c,0] for c in labels1[N_pop]],
        y = [Bias_p[c,PC_select] for c in labels1[N_pop]],
      name= 'PC {}'.format(PC_select - 1)
    ) for PC_select in range(2,Bias_p.shape[1])]


layout = go.Layout(
    title= 'Biased to Unbiased sampling, Pearsons r, N_pops: ' + str(N_pop),
    yaxis=dict(
        title="Pearson's r",
        range= [-.5,1]
    ),
    xaxis=dict(
    title= 'Haplotype length')
)


fig = go.Figure(data=fig_data, layout=layout)
iplot(fig)
```

**Fig. 1** PC-wise comparison of pairwise centroid distances in unbiased versus biased scenarios across haplotype lengths. Number of populations to be selected at the beginning of the block of code.

```python
N_pop= 4

labels1= {z:[x for x in range(MScorr.shape[0]) if MScorr[x,1] == z] for z in list(set(MScorr[:,1]))}

fig_data= [go.Scatter(
        x = [MScorr[c,0] for c in labels1[N_pop]],
        y = [MScorr[c,PC_select] for c in labels1[N_pop]],
      name= 'PC {}'.format(PC_select - 1)
    ) for PC_select in range(2,MScorr.shape[1])]


layout = go.Layout(
    title= 'MS correct to Unbiased sampling, Pearsons r, N_pops: ' + str(N_pop),
    yaxis=dict(
        title= "Pearson's r",
        range= [-.5,1]
    ),
    xaxis=dict(
    title= 'Haplotype length')
)


fig = go.Figure(data=fig_data, layout=layout)
iplot(fig)
```

**Fig. 2** PC-wise comparison of pairwise centroid distances in unbiased versus MS-corrected biased scenarios across haplotype lengths. Number of populations to be selected at the beginning of the block of code.

```python
Schemes= ['unbiased', 'biased', 'corrected']

Select= 'biased'

labels1= {z:[x for x in range(tests.shape[0]) if tests[x,1] == z] for z in list(set(tests[:,1]))}


fig_data= [go.Scatter(
        x = tests[labels1[i],0],
        y = tests[labels1[i],Schemes.index(Select) + 2],
      name= 'N {}, {}'.format(i,round(np.mean(tests[labels1[i],Schemes.index(Select) + 2])*100,1))
    ) for i in labels1.keys()]


layout = go.Layout(
    title= 'Genetic to feature space distances, Sampling scheme: ' + Select,
    yaxis=dict(
        title="Pearsons'r",
        range= [0,1]
    ),
    xaxis=dict(
    title= 'Haplotype length')
)


fig = go.Figure(data=fig_data, layout=layout)
iplot(fig)
```

**Fig. 3** Whole feature space distance comparison between genetic and feature space distances in scenario of choice. Scenario to be selected at the begining o the block, interactive graphics are not yet supported by NBviewer.
